```python
def notices(soup):
    
    featured_article = soup.find('div', attrs={
        'class': 'featured-article__container'
    })
    featured_article_link = [featured_article.find('a').get('href')]
    
    articles_h2 = soup.find('ul', attrs={
        'class': 'article-list'
    }).find_all('h2')
    articles_a = [article_h2.find('a', attrs= {
        'class': ''
    }) for article_h2 in articles_h2]
    
    articles_links = [article_a.get('href') for article_a in articles_a]
    
    return featured_article_link + articles_links
```

## Descargando contenido

```python
url_nota = 'https://www.pagina12.com.ar/233660-clonacion-bebes-de-diseno-etica-y-tecnologia'
```

```python
def extract_info(url_nota):
    """Función que extrae la información de una nota a partir de su url"""
    try:
        ret_dict = {}
        nota = requests.get(url_nota)
        if nota.status_code == 200:
            s_nota = BeautifulSoup(nota.text, 'lxml')

            # Extraemos el título
            title = s_nota.find('div', attrs={
                'class': 'article-title'
            }).get_text()
            # print('Título:', title, '\n')
            ret_dict['title']= title

            # Extraemos la fecha
            date = s_nota.find('span', attrs={
                'pubdate': 'pubdate'
            }).get('datetime')
            # print('Fecha:', date, '\n')
            ret_dict['date']= date

            # Extraemos el copete
            try:
                summary = s_nota.find('div', attrs={
                    'class': 'article-summary'
                }).get_text()
                # print('Copete:', summary, '\n')
                ret_dict['summary']= summary
            except:
                # print('Copete: vacío', '\n')
                ret_dict['summary']= None

            # Extraemos la volanta
            try:
                prefix = s_nota.find('div', attrs={
                    'class': 'article-prefix'
                }).get_text()
                # print('Volanta:', prefix, '\n')
                ret_dict['prefix']= prefix
            except:
                # print('Volanta: vacío')
                ret_dict['prefix']= None
            
            # Extraemos las imágenes
            try:
                media = s_nota.find('div', attrs={
                    'class': 'article-main-media-image'
                }).find_all('img')
                if len(media) == 0:
                    # print('Media: vacío', '\n')
                    ret_dict['media']= None
                else:
                    image_link = media[-1].get('data-src')
                    
                    # Descargamos la imagen
                    img_req = requests.get(image_link)
                    
                    # Mostramos la imagen
                    # from IPython.display import Image, display
                    # print('Media:', '\n')
                    # display(Image(img_req.content))
                    ret_dict['media']= img_req
            except:
                 # print('Media: vacío', '\n')
                 ret_dict['media']= None
            
            # Extraemos el cuerpo
            body_paragraphs = s_nota.find('div', attrs={
                'class': 'article-body'
            }).find('div', attrs={
                'class': 'article-text'
            }).find_all('p')

            body = "\n\n".join([p.text for p in body_paragraphs])
            # print('Cuerpo:', body, '\n')
            ret_dict['body']= body

            #Extraemos el autor
            try:
                author = s_nota.find('div', attrs={
                    'class': 'article-author'
                }).span.a.get_text()
                # print('Autor:', author)
                ret_dict['author']= author
            except:
                # print('Autor: anónimo')
                ret_dict['author']= None
            
            return ret_dict

    except Exception as e:
        print('error:', e, '\n')
        return ret_dict
```

```python
extract_info('https://www.pagina12.com.ar/234350-el-costo-oculto-de-ir-a-estudiar')
```

## Unificando el scraper

```python
def scrape_nota(url):
    
    ret_dict = extract_info(url)
    ret_dict['url']: url
    return ret_dict
```

```python
notas = []
```

```python
for link in links_secciones:
    try:
        r = requests.get(link)
        if r.status_code == 200:
            soup = BeautifulSoup(r.text, 'lxml')
            notas.extend(notices(soup))
        else:
            print('No se pudo obtener la sección:', link)
    except Exception as e:
            print('Error:', e)
```

```python
notas
```

```python
len(notas)
```

```python
data = []
```

```python
for i, nota in enumerate(notas):
    print(f'Scrapeando nota {i}/{len(notas)}')
    data.append(scrape_nota(nota))
```
