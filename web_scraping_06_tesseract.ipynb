```python
import tesserocr # Para hacer OCR
import numpy as np # Para hacer manipulación básica de imágenes
import matplotlib.pyplot as plt # Para visualizar imágenes
from PIL import Image # Para cambiar el formato de archivos
%matplotlib inline
```

```python
texto_largo = plt.imread('texto_largo.jpg')
plt.figure(figsize=(15,5))
plt.imshow(texto_largo)
plt.axis(False)
```

```python
texto_ocr = tesserocr.file_to_text('texto_largo.jpg', lang='spa')
```

```python
texto_ocr.replace('\n', '')
```

```python
img_rgb = img[:,:,:3]
```

```python
img_rgb.shape
```

```python
plt.imshow(img_rgb)
```

```python
# Invertir tonos

img_rgb[0,0,0]
```

```python
img_inv = 1 - img_rgb
```

```python
plt.imshow(img_inv)
```

```python
# Imagen en escala de grises

img_rg = img_inv.mean(axis=2)
```

```python
plt.imshow(img_rg, cmap='Greys_r')
```

```python
tesserocr.image_to_text(img_rg, lang='spa')
```

```python
img_pil = Image.fromarray(np.uint8(img_rg*255))
```

```python
type(img_pil)
```
